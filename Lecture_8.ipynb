```python
earthquakes = pd.read_csv("earthquakes.csv")
# earthquakes['date'] = pd.to_datetime(earthquakes['date'])

display(earthquakes.head())
```

```python
display_dataframe(earthquakes)
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=earthquakes, x='Year', y='earthquakes_per_year')

plt.title("Time-series plot of the earthquakes per year")
plt.xlabel("Year")
plt.xticks(np.arange(earthquakes['Year'].min(), earthquakes['Year'].max()+5, 5), rotation=90)
plt.yticks(np.arange(0, earthquakes['earthquakes_per_year'].max()+10, 5))

# plt.xticks(np.arange(1900, 2003, 5))
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(earthquakes['Year'], earthquakes['earthquakes_per_year'], linestyle='--')
plt.scatter(earthquakes['Year'], earthquakes['earthquakes_per_year'], marker='o', color='r')

# plt.fill_between(earthquakes['Year'], earthquakes['earthquakes_per_year'], color='b', alpha=0.25)
plt.fill_between(earthquakes['Year'], 20, 30, color='b', alpha=0.25)

plt.title("Time-series plot of the earthquakes per year")
plt.xlabel("")
plt.ylabel("earthquakes_per_year")
plt.xticks(np.arange(earthquakes['Year'].min(), earthquakes['Year'].max()+5, 5), rotation=90)
plt.yticks(np.arange(0, earthquakes['earthquakes_per_year'].max()+10, 5))

plt.show()
```
