```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('talk')
```

#### Creating new dataframe

```python
def straight_line(x, m, c):
    
    df = pd.DataFrame()

    df['x'] = x
    df['y'] = m*x + c #y=mx+c
    
    return df
```

#### Creating new straight lines

```python
SL_1 = straight_line(np.arange(-10,11,1), 1, 1)
SL_2 = straight_line(np.arange(-10,11,1), 1, -1)
SL_3 = straight_line(np.arange(-10,11,1), -1, 1)
SL_4 = straight_line(np.arange(-10,11,1), -1, -1)
SL_5 = straight_line(np.arange(-10,11,1), 1, 0)
SL_6 = straight_line(np.arange(-10,11,1), -1, 0)
SL_7 = straight_line(np.arange(-10,11,1), 0, 5)
SL_8 = straight_line(np.arange(-10,11,1), 0, -5)
SL_9 = straight_line(np.arange(-10,11,1), 0, 0)
```

#### Create a lineplot function

```python
def lineplot(x_axis, y_axis, title):
    plt.plot(x_axis, y_axis, marker='.', color='red')
    plt.vlines(0, -15, 15, linestyles=':', colors="black")
    plt.hlines(0, -10, 10, linestyles=':', colors="black")
    plt.xticks(np.arange(-10,11,5))
    plt.yticks(np.arange(-15,16,5))
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
```

#### Creating subplots

```python
plt.figure(figsize=(20,12))

plt.subplot(3,3,1)
lineplot(SL_1['x'], SL_1['y'], "y=mx+c, where m=1, c=1")

plt.subplot(3,3,2)
lineplot(SL_2['x'], SL_2['y'], "y=mx-c, where m=1, c=-1")

plt.subplot(3,3,3)
lineplot(SL_3['x'], SL_3['y'], "y=-mx+c, where m=-1, c=1")

plt.subplot(3,3,4)
lineplot(SL_4['x'], SL_4['y'], "y=-mx-c, where m=-1, c=-1")

plt.subplot(3,3,5)
lineplot(SL_5['x'], SL_5['y'], "y=mx, where m=1, c=0")

plt.subplot(3,3,6)
lineplot(SL_6['x'], SL_6['y'], "y=-mx, where m=-1, c=0")

plt.subplot(3,3,7)
lineplot(SL_7['x'], SL_7['y'], "y=c, where m=0, c=5")

plt.subplot(3,3,8)
lineplot(SL_8['x'], SL_8['y'], "y=-c, where m=0, c=-5")

plt.subplot(3,3,9)
lineplot(SL_9['x'], SL_9['y'], "y=0, where m=0, c=0")

plt.tight_layout()
plt.show()
```

#### Varying the marker, color, and linestyle

```python
plt.figure(figsize=(20,12))

plt.subplot(3,3,1)
plt.plot(SL_1['x'], SL_1['y'], marker='.', color='red')
plt.title("row = 1, col = 1")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,2)
plt.plot(SL_1['x'], SL_1['y'], marker='o', color='red')
plt.title("row = 1, col = 2")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,3)
plt.plot(SL_1['x'], SL_1['y'], marker='s', color='red')
plt.title("row = 1, col = 3")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,4)
plt.plot(SL_1['x'], SL_1['y'], color='red', linestyle="--")
plt.title("row = 2, col = 1")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,5)
plt.plot(SL_1['x'], SL_1['y'], color='red', linestyle="-.")
plt.title("row = 2, col = 2")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,6)
plt.plot(SL_1['x'], SL_1['y'], color='red', linestyle=":")
plt.title("row = 2, col = 3")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,7)
plt.plot(SL_1['x'], SL_1['y'], color='red', alpha=0.5)
plt.title("row = 3, col = 1")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,8)
plt.plot(SL_1['x'], SL_1['y'], color='red')
plt.title("row = 3, col = 2")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,3,9)
plt.plot(SL_1['x'], SL_1['y'], color='b')
plt.title("row = 3, col = 3")
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

#### Multiple lineplots in the same figure

```python
plt.figure(figsize=(20,12))

plt.plot(SL_1['x'], SL_1['y'], color='red', label='m=1, c=1')
plt.plot(SL_2['x'], SL_2['y'], color='green', label='m=1, c=-1')
plt.plot(SL_3['x'], SL_3['y'], color='blue', label='m=-1, c=1')
plt.plot(SL_4['x'], SL_4['y'], color='orange', label='m=-1, c=-1')
plt.vlines(0, -15, 15, linestyles=':', colors="black")
plt.hlines(0, -10, 10, linestyles=':', colors="black")
plt.title("y=mx+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20,12))

plt.plot(SL_5['x'], SL_5['y'], color='red', label='m=1, c=0')
plt.plot(SL_6['x'], SL_6['y'], color='green', label='m=-1, c=0')
plt.plot(SL_7['x'], SL_7['y'], color='blue', label='m=0, c=5')
plt.plot(SL_8['x'], SL_8['y'], color='orange', label='m=0, c=0')
plt.plot(SL_9['x'], SL_9['y'], color='m', label='m=0, c=-5')
plt.vlines(0, -15, 15, linestyles=':', colors="black")
plt.hlines(0, -10, 10, linestyles=':', colors="black")
plt.title("y=mx+c")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.show()
```
